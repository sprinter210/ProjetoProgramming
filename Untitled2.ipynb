```python
## Definir uma função chamada "loss_function" que que calcula o erro quadrático médio entre a variável "salary" 
# e os valores preditos usando os parâmetros da regressão linear.
#

def loss_function(m, b, points):
    total_error = 0
    for i in range(len(points)):
        x = points.iloc[i].age
        y = points.iloc[i].salary
        total_error += (y - (m * x + b)) ** 2
    total_error / float(len(points))
```

```python
def gradient_descent(m_now, b_now, points, L):
    m_gradient = 0
    b_gradient = 0

    n = len(points)

    for i in range(n):
        x = points.iloc[i].age
        y = points.iloc[i].salary

        m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
        b_gradient += (-2/n) * (y - (m_now * x + b_now))

    m = m_now - m_gradient * L
    b = b_now - b_gradient * L
    return m, b
```

```python
m=0
b=0
L = 0.0001
epochs = 1000
```

```python
for i in range(epochs):
    if i % 25 == 0:
        print(f"Epoch: {i}")
    m, b = gradient_descent(m, b, data, L)
```
